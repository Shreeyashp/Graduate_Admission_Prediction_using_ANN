```python
df.drop(columns = ['Serial No.'], inplace = True)
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation='relu',input_dim=7)) #input_dim=7 because 7 column, hiddenlayer1 will have 7 nodes
model.add(Dense(1,activation='linear')) # linear because model is of regression problem
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='mean_squared_error') #loss='mean_squared_error' for linear
```

```python
history = model.fit(X_train_scaled,y_train,epochs=10,validation_split = 0.2)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

> there is need to improve model accuracy

```python
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split = 0.2)
```

```python
y_pred = model.predict(X_test_scaled)
```
